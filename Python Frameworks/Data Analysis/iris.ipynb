```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.express as px
sns.set(style="white", color_codes=True)

iris = pd.read_csv("/Users/greenboi/ComputerScience/Python/Python Projects and Frameworks/Data Analysis/Data/Iris Data/Iris.csv")

iris.head()
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
px.scatter(iris, x="SepalLengthCm", y="SepalWidthCm")
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
```

```python
sns.FacetGrid(iris, hue="Species") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
px.scatter(iris, x="SepalLengthCm", y="SepalWidthCm", color="Species")
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
px.box(iris, x="Species", y="PetalLengthCm", color="Species")
```

```python
px.violin(iris, x="Species", y="PetalLengthCm", color="Species", box=True)
```

```python
sns.FacetGrid(iris, hue="Species") \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation between each pair of features
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
```

```python
# Now that we've covered seaborn, let's go back to some of the ones we can make with Pandas
# We can quickly make a boxplot with Pandas on each feature split out by species
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```

```python
# One cool more sophisticated technique pandas has available is called Andrews Curves
# Andrews Curves involve using attributes of samples as coefficients for Fourier series
# and then plotting these
from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
```

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

```python
from pandas.plotting import radviz 
radviz(iris.drop("Id", axis=1), "Species")
```
